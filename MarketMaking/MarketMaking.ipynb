```python
import time
import random
from typing import Dict, Tuple

class MarketMakingBot:
    def __init__(self, initial_price: float, spread: float, inventory_limit: int):
        self.current_price = initial_price
        self.spread = spread  # Difference between bid and ask prices
        self.inventory = 0    # Current position
        self.inventory_limit = inventory_limit
        self.cash = 10000     # Starting cash
        self.order_book: Dict[str, Tuple[float, int]] = {
            'bid': (0, 0),   # (price, quantity)
            'ask': (0, 0)
        }

    def update_prices(self) -> None:
        """Update bid and ask prices based on current market price"""
        # Simulate market price movement
        price_change = random.uniform(-0.5, 0.5)
        self.current_price += price_change
        
        # Calculate bid and ask prices
        bid_price = self.current_price - (self.spread / 2)
        ask_price = self.current_price + (self.spread / 2)
        
        # Adjust quantities based on inventory
        bid_qty = max(1, int(10 - self.inventory))
        ask_qty = max(1, int(10 + self.inventory))
        
        # Update order book
        self.order_book['bid'] = (bid_price, bid_qty)
        self.order_book['ask'] = (ask_price, ask_qty)

    def simulate_trade(self) -> None:
        """Simulate random market participants interacting with our quotes"""
        if random.random() < 0.5:  # 50% chance of a trade
            if random.random() < 0.5:  # Buy from our ask
                if self.inventory > -self.inventory_limit:
                    qty = min(self.order_book['ask'][1], random.randint(1, 3))
                    self.inventory -= qty
                    self.cash += qty * self.order_book['ask'][0]
                    print(f"Market bought {qty} @ {self.order_book['ask'][0]:.2f}")
            else:  # Sell to our bid
                if self.inventory < self.inventory_limit:
                    qty = min(self.order_book['bid'][1], random.randint(1, 3))
                    self.inventory += qty
                    self.cash -= qty * self.order_book['bid'][0]
                    print(f"Market sold {qty} @ {self.order_book['bid'][0]:.2f}")

    def run(self, iterations: int) -> None:
        """Run the market making simulation"""
        print(f"Starting with ${self.cash} cash and inventory {self.inventory}")
        
        for i in range(iterations):
            self.update_prices()
            print(f"\nTick {i+1}:")
            print(f"Bid: {self.order_book['bid'][0]:.2f} x {self.order_book['bid'][1]} | "
                  f"Ask: {self.order_book['ask'][0]:.2f} x {self.order_book['ask'][1]}")
            
            self.simulate_trade()
            print(f"Inventory: {self.inventory} | Cash: ${self.cash:.2f}")
            
            if abs(self.inventory) >= self.inventory_limit:
                print("Inventory limit reached!")
                break
                
            time.sleep(1)  # Simulate time between ticks

if __name__ == "__main__":
    # Initialize bot with starting price of 100, spread of 0.2, and inventory limit of 20
    bot = MarketMakingBot(initial_price=100.0, spread=0.2, inventory_limit=20)
    bot.run(iterations=10)  # Run for 10 iterations
```
